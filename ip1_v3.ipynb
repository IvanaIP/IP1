## Min‚ÄìMax normalizacija
Formula:

$\Large x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$



Transformisemo vrednosti atributa iz prirodnog raspona [min, max], na **raspon [0, 1]**.

```python
def minmax(X):
      return (X - min(X))/(max(X) - min(X))
```

# Odsecanje

Ako podaci imaju **ekstremne autlajere** mo≈æemo koristiti odsecanje.<br> 
Tada se sve vrednosti manje od minimalne zadate slikaju u min. Analogno va≈æi za max.

<div style="text-align:left; font-size: 1.em;">
\(\displaystyle
\text{clipping}(x, \min, \max) =
\begin{cases}
\min, & x < \min \\
\max, & x > \max \\
\dfrac{x - \min}{\max - \min}, & \text{inaƒçe}
\end{cases}
\)
</div>

```python
# Mo≈æe se primeniti pre i posle drugih vrsta normalizacije.
def clipping(X, min, max):
    X.where(X > min, min, inplace=True)
    X.where(X < max, max, inplace=True)
    return X
```

```python
titanic_clip = clipping(titanic['fare'], titanic['age'].min(), titanic['age'].max())
titanic_clip
```

## Z-score (standardizacija)
Z-score predstavlja koliko standardinh devijacija je vrednost udaljena od srednje vrednosti.

Koristimo kada ≈æelimo da osiguramo da je mean = 0 i std = 1. Korisno ako imamo outlajere, ali ne ekstremne da nam treba odsecanje

$x' = \frac{x - \mu}{\sigma}$

Œº = srednja vrednost

œÉ = standardna devijacija


```python
def z_score(X):
    return (X - X.mean()) / X.std()
```
